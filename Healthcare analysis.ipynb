```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

# Checking the outliers

```python
columns = ["Age" ,"Billing Amount"]

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1

    LB = Q1-1.5*IQR
    UB = Q3+1.5*IQR

    outlier = (df[col]<LB) | (df[col]>UB)
    print(outlier.sum())
```

# Distribution of age and charges.


```python
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
sns.histplot(data = df, x = 'Age' ,kde=True, color = 'blue')
plt.title("Age Distribution")

plt.subplot(1,2,2)
sns.histplot(data = df, x = 'Billing Amount' ,kde=True, color = 'green')
plt.title("Billing Amount Distribution")
```

# Bivariate Analysis

# Compare average charges by gender.

```python
df.groupby('Gender')['Billing Amount'].mean().round(2).sort_values(ascending = False)
```

```python
t = df.groupby('Gender')['Billing Amount'].mean().round(2).sort_values(ascending = False).plot(kind = "bar" , color = "orange")
for bar in t.containers:
    t.bar_label(bar)
plt.title("Average Charges by Gender")
```

# Compare average charges by diagnosis.

```python
df.groupby('Medical Condition')['Billing Amount'].mean().round(2).sort_values(ascending = False)
```

```python
t = df.groupby('Medical Condition')['Billing Amount'].mean().round(2).sort_values(ascending = False).plot(kind = "bar" , color = "green")
for bar in t.containers:
    t.bar_label(bar)
plt.title("Average Charges by Medical Condition")
```

# Relationship between patient age and hospital charges

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Age", y="Billing Amount", hue="Test Results",
                s=30)   
plt.title("Age vs Billing Amount by Test Results")
plt.show()
```

# Checking whether older patients tend to have longer hospital stays.

```python
df['Total Stay(in days)'] = df['Discharge Date']- df['Date of Admission']
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data = df , x = "Age", y ="Total Stay(in days)")
plt.title("Age vs Length of Hospital Stay")
plt.xlabel("Patient Age")
plt.ylabel("Hospital Stay (days)")
plt.show()

```

# total and average hospital charges by diagnosis.

```python
df.groupby('Medical Condition')['Billing Amount'].agg(["sum", "mean"])
```

```python
plt.figure(figsize = (10,3))
plt.subplot(1,2,1)
b = df.groupby('Medical Condition')['Billing Amount'].sum().sort_values(ascending = False).plot(kind = "bar", color = 'blue')
plt.title("Total Charges by Diagnosis")

plt.subplot(1,2,2)
df.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending = False).plot(kind = "bar", color = 'orange')
plt.title("Average Charges by Diagnosis")
```
